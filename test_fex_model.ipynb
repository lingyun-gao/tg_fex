```python
import numpy as np

from __init__ import FeatureExtractorModelVersioned
from tg_utils import get_pad_image_from_url
```

```python
# TODO: set this to a high number, maybre 64 or 128 on GPU machine
batch_size = 2
```

```python
# load model

m = FeatureExtractorModelVersioned(batch_size=batch_size)
```

```python
# create a sample batch

url1 = 'https://www.gotceleb.com/wp-content/uploads/photos/dakota-johnson/in-red-t-shirt-out-in-new-york/Dakota-Johnson-in-Red-T-Shirt--08-662x955.jpg'
url2 = 'http://s4x9c3w4.stackpathcdn.com/wp-content/uploads/2014/09/P1014511.jpg'

img1 = get_pad_image_from_url(url1)
img2 = get_pad_image_from_url(url2)

b1 = m.t.transform(img1)
b2 = m.t.transform(img2)

batch_in = np.array([b1, b2])
```

```python
# now run forward pass on the batch

m.forward_batch(batch_in)
```

```python
m.t.init_batch(3)
```
